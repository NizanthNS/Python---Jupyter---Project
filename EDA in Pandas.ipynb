```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()
corr
```

```python
plt.figure(figsize=(10, 8))
sb.heatmap(corr, annot = True)

plt.show()
```

```python
df.columns
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns

population_cols = [
    '1970 Population',
    '1980 Population',
    '1990 Population',
    '2000 Population',
    '2010 Population',
    '2015 Population',
    '2020 Population',
    '2022 Population'
]

population_cols = [col for col in population_cols if col in numeric_cols]

df2 = (
    df.groupby('Continent')[population_cols]
      .mean()
      .sort_values(by='2022 Population', ascending=False)
)
df2
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df3 = df2.transpose()
df3
```
