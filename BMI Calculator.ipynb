# BMI Calculator

```python
# BMI = weight (kg) / [height (m)]Â²
```

```python
Classification	BMI range - kg/m2
Severe Thinness	< 16
Moderate Thinness	16 - 17
Mild Thinness	17 - 18.5
Normal	18.5 - 25
Overweight	25 - 30
Obese Class I	30 - 35
Obese Class II	35 - 40
Obese Class III	> 40
```

```python
Name = input("Enter your Name: ")

Weight = int(input("Enter your Weight in Kilogram : "))

Height = float(input("Enter your Height in Meter : "))

BMI = (Weight / Height**2)

print(BMI)

if BMI>0:
    if(BMI<16):
        print(Name + ', You are Severely Thin. You are underweight and may need extra nutrition with care, you can regain strength.')
    elif(BMI<17):
        print(Name +', You are Moderately Thin. You are a bit underweight, but healthy eating can help you reach balance.')
    elif(BMI<18.5):
        print(Name +', You are Mildly Thin. Slightly below normal, a few lifestyle changes can make a big difference.')
    elif(BMI<25):
        print(Name +', You are in the Normal Weight range. You are in a healthy weight range, keep maintaining your balance.')
    elif(BMI<30):
        print(Name +', You are Overweight. You are a bit above normal, small and steady changes can improve your health.')
    elif(BMI<35):
        print(Name +', You are in Obese Class I. You are in the first obesity range, focus on gradual and healthy habits.')
    elif(BMI<40):
        print(Name +', You are in Obese Class II. Your BMI is high, with support and consistency, improvement is possible.')
    elif(BMI>40):
        print(Name +', You are in Obese Class III. Your BMI is very high, take it one step at a time and seek support for better health.')
    else:
        print('Enter valid inputs')
```
