# File Sorter

```python
import os
import shutil
```

```python
path = r"D:/Data Analyst/Python/Files/"
```

```python
print(os.listdir(path))
```

```python
folder_names = ['xlsx files', 'JPG files', 'PNG files', 'Jpeg files', 'Text']
```

```python
for folder in folder_names:
    folder_path = os.path.join(path, folder)
    if not os.path.exists(folder_path):
        print(f"Creating folder: {folder_path}")
        os.makedirs(folder_path)
```

```python
for file in os.listdir(path):
    if file.endswith('.xlsx'):
        shutil.move(os.path.join(path, file), os.path.join(path, 'xlsx files', file))
    elif file.lower().endswith('.jpg'):
        shutil.move(os.path.join(path, file), os.path.join(path, 'JPG files', file))
    elif file.lower().endswith('.jpeg'):
        shutil.move(os.path.join(path, file), os.path.join(path, 'Jpeg files', file))
    elif file.lower().endswith('.png'):
        shutil.move(os.path.join(path, file), os.path.join(path, 'PNG files', file))
    elif file.lower().endswith('.txt'):
        shutil.move(os.path.join(path, file), os.path.join(path, 'Text', file))
```
